```python
import pandas as pd
import numpy as np
```

```python
demand_data = pd.read_csv("abs_data.csv")
help_centres = pd.read_csv("help_centres.csv")
sensors = pd.read_csv("sensors_sat_11am_with_postcodes.csv")
```

```python
demand_data.head()
```

```python
help_centres.head()
```

```python
sensors["postcode"] = np.round(sensors["mccid_int"]).astype(int) # convert postcodes to int
```

```python
sensors.head()
```

```python
demand_data["demand"] = demand_data["$1 to 1000"] + demand_data["1000-1999"]
```

```python
demand_data
```

```python
demand_years = demand_data[["Postcode", "Income year", "demand"]]
demand_years
```

```python
demand_years = demand_data.pivot(index="Postcode", columns="Income year", values="demand").reset_index()
demand_years
```

```python
demand_years.columns
```

```python
# 2021 projection. y = mx + c.
demand_years["projected_demand"] = demand_years[2016] + (demand_years[2016] - demand_years[2006])
```

```python
demand_years
```

```python
sensors.groupby(['postcode']).count()
```

```python
sensors.groupby(['postcode', 'status'])['status'].count()
```

```python
present = sensors[sensors['status'] == 'Present'].groupby(['postcode'])['postcode'].count()
present = present.rename('present')
present
```

```python
unoccupied = sensors[sensors['status'] == 'Unoccupied'].groupby(['postcode'])['postcode'].count()
unoccupied = unoccupied.rename('unoccupied')
unoccupied
```

```python
occupancy = pd.concat([present, unoccupied], axis=1).reset_index()
occupancy
```

```python
occupancy["pct_free"] = occupancy["unoccupied"] / (occupancy["present"] + occupancy["unoccupied"])
```

```python
occupancy
```

```python
occupancy["postcode"]
```

```python
merge1 = pd.merge(occupancy, demand_years, left_on="postcode", right_on="Postcode", how='left')
merge1
```

```python
merge2 = pd.merge(merge1, help_centres, left_on="postcode", right_on="Post Code", how='left')
merge2
```
